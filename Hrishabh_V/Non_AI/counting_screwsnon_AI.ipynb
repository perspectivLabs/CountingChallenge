```python
import cv2
from google.colab.patches import cv2_imshow

def count_objects(image_path):
    img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)#Greyscale
    ret, thresh = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY_INV)#binary threshold
    contours, _ = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)#contours
    num_Screws = len(contours)
    img_color = cv2.imread(image_path)
    cv2.drawContours(img_color, contours, -1, (0, 255, 0), 2)
    cv2_imshow(img_color)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    return num_objects

if __name__ == "__main__":
    image_path = '/content/drive/MyDrive/Perspective labs/Screws_2024_07_15-20240731T083126Z-001/Screws_2024_07_15/img1.jpg'  #path to your image file
    num_Screws = count_objects(image_path)
    print(f"Number of Screws and bolts: {num_Screws}")
```
